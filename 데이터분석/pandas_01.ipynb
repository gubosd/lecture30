### DataFrame 을 numpy 데이터로 변환
- 주로 숫자부분만을 numpy 로 변환한다

```python
a = subway.iloc[:, 4:6].values
a
```

```python
a = subway.iloc[:, 4:6].to_numpy()
a
```

```python
plt.scatter(a[:, 0], a[:, 1], alpha=0.1) # alpha 는 투명도
```

- numpy 로 변환시 데이터타입에 주의해야 한다 (정수타입인지 실수타입인지)

```python
a.dtype # 모든 데이터가 정수값이므로 정수형 어레이가 된다
```

```python
b = a.astype('float') # 머신러닝에서는 실수가 아니면 에러가 발생할 수 있다
b
```

### 칼럼 추가하기
- 새로운 칼럼명에 값을 지정하면 칼럼이 추가됨

```python
subway['승객수평균'] = (subway.승차총승객수 + subway.하차총승객수)/2
subway
```

```python
a = subway[['승차총승객수', '하차총승객수', '승객수평균']].values
a           # 실수값이 섞여 있으므로 실수형 어레이로 변환된다
```
