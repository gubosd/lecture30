### map(), apply(), applymap()

```python
subway.노선명.map(lambda s: '<' + s + '>')
```

```python
subway[['승차총승객수','하차총승객수']].apply(lambda series: series.max()-series.min()) # 열단위로 처리
```

```python
subway[['승차총승객수','하차총승객수']].applymap(lambda x: x//1000) # 각 항목에 적용
```

### 그룹 연산
- groupby() 한 후 sum(), mean(), std(), count(), size() 등 적용

```python
g1 = subway.groupby('사용일자')[['승차총승객수','하차총승객수']].sum()
g1
```

```python
g1.shape
```

```python
plt.plot(g1['승차총승객수'].values)
plt.xticks(range(g1.shape[0]), g1.index, rotation=90)
pass
```

```python
g2 = subway.groupby(['사용일자','노선명'])['승차총승객수'].sum()
g2
```

```python
g2.unstack()
```

```python
t = pd.to_datetime(subway.사용일자, format='%Y%m%d')
subway['요일'] = t.dt.dayofweek
subway
```

```python
g3 = subway.groupby(['요일','노선명'])['승차총승객수'].sum()
g3
```

```python
result = g3.unstack()
result
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 적용
plt.plot(result.values[:,:4], '^-')
plt.legend(['1호선','2호선','3호선','4호선'])
plt.xticks(range(7), list('월화수목금토일'))
pass
```

```python
subway.groupby('노선명')['승차총승객수'].agg(['sum','mean','std','count','size'])
```
